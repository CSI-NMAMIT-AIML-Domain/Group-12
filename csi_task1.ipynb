```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
des=df.sort_values(by='confirmed',ascending=False).head(10)

sns.barplot(x="country",y="confirmed",data=des)
plt.title("Confirmed cases of Top 10 countries")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['confirmed'], df['deaths'])
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.title('Confirmed Cases vs Deaths')
plt.show()
```

```python
sns.displot(x="recovered",y="country",data=des)
plt.title("Recovery of Top 10 countries")
```
